```
import pandas as pd
import numpy as np
import re
```

```
!curl https://data.cityofnewyork.us/api/views/rc75-m7u3/rows.csv?accessType=DOWNLOAD -o data.csv

df = pd.read_csv("data.csv", parse_dates=True, index_col='DATE_OF_INTEREST')
df
```

```
#Use column selection to see total case count across the five boroughs and city totals


col_borough_cases = (['BX_CASE_COUNT','BK_CASE_COUNT','MN_CASE_COUNT','QN_CASE_COUNT','SI_CASE_COUNT'])
city_totals = (['CASE_COUNT','HOSPITALIZED_COUNT','DEATH_COUNT'])

```

```
#Case, Hospitalization, and Death counts per day throughout the New York City

df_city_totals = df[city_totals]

df_city_totals
```

```
#Case, Hospitalization, and Death Total for COVID-19 throughout NYC


df_city_totals.sum()
```

```
#Percentage of people hospitalized for those who tested positive for COVID-19 throughout NYC

hospToCase = ((df_city_totals['HOSPITALIZED_COUNT'].sum()/df_city_totals['CASE_COUNT'].sum())*100)

hospToCase = str(hospToCase)
hospToCase = re.sub(r'\.\d*', '%', hospToCase)

hospToCase
```

```
#Percentage of people who died for those who tested positive for COVID-19 throughout NYC

deathToCase = ((df_city_totals['DEATH_COUNT'].sum()/df_city_totals['CASE_COUNT'].sum())*100)

deathToCase = str(deathToCase)
deathToCase = re.sub(r'\.\d*', '%', deathToCase)

deathToCase
```

```
#Percentage of deaths for those who entered the hospital for COVID-19 throughout NYC

deathToHosp = ((df_city_totals['DEATH_COUNT'].sum()/df_city_totals['HOSPITALIZED_COUNT'].sum())*100)

deathToHosp = str(deathToHosp)
deathToHosp = re.sub(r'\.\d*', '%', deathToHosp)

deathToHosp
```

```
col_borough_deaths = (['BX_DEATH_COUNT','BK_DEATH_COUNT','MN_DEATH_COUNT','QN_DEATH_COUNT','SI_DEATH_COUNT'])

df_borough_deaths=df[col_borough_deaths]

df_borough_deaths                                            
```

```
caseWinter = np.array([])

for i in df_borough_case_winter:
  caseWinter = np.append(caseWinter, i)

```

```
hospWinter = np.array([])

for i in df_borough_hosp_winter:
  hospWinter = np.append(hospWinter, i)
hospWinter
```

```
boroughList = ['BRONX', 'BROOKLYN', 'MANHATTAN', 'QUEENS', 'STATEN ISLAND']
k=0
print('Percentage of people hospitalized who were diagnosed with COVID-19 during the beginning of the pandemic until Spring:\n')
for i in caseToHospWinter:
  x = i*100
  j=str(x)
  j = re.sub(r'\.\d*', '%', j)
  print(boroughList[k], " : ", j)
  k+=1
```
